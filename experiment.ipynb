```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
## Load the dataset

data = pd.read_csv("Churn_Modelling.csv")
data.head()
```

```python
###Our next Step is the preprocessing the data 

data = data.drop(['RowNumber','CustomerId','Surname'],axis=1)
data
```

```python
## Encoding categorial Values as Gender to 1 and 0's

label_encoder_gender = LabelEncoder()
data['Gender'] = label_encoder_gender.fit_transform(data['Gender'])
data
```

```python
## Onehot encode Geography 

from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo = OneHotEncoder()
geo_encoder = onehot_encoder_geo.fit_transform(data[['Geography']])
geo_encoder
```

```python
onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoder_df=pd.DataFrame(geo_encoder.toarray(),columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoder_df
```

```python
### Combining One Hot Encoder columns with our Original Data 

data = pd.concat([data.drop('Geography',axis=1),geo_encoder_df],axis=1)
data.head()
```

```python
### Combining One Hot Encoder columns with our Original Data 

if 'Geography' in data.columns:
    data = pd.concat([data.drop('Geography', axis=1), geo_encoder_df], axis=1)
    
data.head()
```

```python
## Save the encoder and scaler 

with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)

with open('onehot_encoder_geo.pkl','wb') as file :
    pickle.dump(onehot_encoder_geo,file)   
```

```python
data.head()
```

```python
## Divide the Dataset into independet and dependent features

x = data.drop('Exited',axis=1)
y = data['Exited']

## Split the data in Training and Testing Sets

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=43)

## SCale these features 

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
with open ('scaler.pkl','wb') as file :
    pickle.dump(scaler,file)
```

ANN Implementation

```python
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
## Build our ANN model

model = Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),  ## HL1 connected with I/P layer
    Dense(32,activation='relu'), ## HL2
    Dense(1,activation='sigmoid') ## O/P layer
])
```

```python
model.summary()
```

```python
import tensorflow

opt = tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss = tensorflow.keras.losses.BinaryCrossentropy()
loss
```

```python
## Compile the model

model.compile(optimizer=opt,loss="binary_crossentropy",metrics=['accuracy'])
```

```python
## Set up the Tensorboard 

from tensorflow.keras.callbacks import EarlyStopping,TensorBoard

log_dir = "log/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback = TensorBoard()
```

```python
### Set up Early Stopping 

early_stopping_callback = EarlyStopping(monitor='val_loss',patience=15,restore_best_weights=True)
```

```python
## Train the Model

history = model.fit(
    x_train,y_train,validation_data= (x_test,y_test),epochs=100,
    callbacks = [tensorflow_callback,early_stopping_callback]
    
)
```

```python
model.save('model.h5')
```
