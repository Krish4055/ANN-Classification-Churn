```python
import tensorflow
from tensorflow.keras.models import load_model 
import pickle
import numpy as np
import pandas as pd 
```

```python
### load the trained model ,scaler pickle ,OHE

model = load_model('model.h5')


## load the encoder and scaler 

with open('onehot_encoder_geo.pkl','rb') as file :
    label_encoder_geo=pickle.load(file)

with open('label_encoder_gender.pkl','rb') as file :
    label_encoder_gender=pickle.load(file)

with open('scaler.pkl','rb') as file :
    scaler=pickle.load(file)
```

```python
input_data ={
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts': 2 ,
    'HasCrCard':1 ,
    'IsActiveMember':1,
    'EstimatedSalary':50000
    }
```

```python
### One hot encode 'Geography'

geo_encode = label_encoder_geo.transform([[input_data['Geography']]]).toarray()
geo_encode_df = pd.DataFrame(geo_encode,columns= label_encoder_geo.get_feature_names_out(['Geography']))
geo_encode_df
```

```python
geo_encode = label_encoder_geo.transform([[input_data['Geography']]]).toarray()
geo_encode_df = pd.DataFrame(geo_encode, columns=label_encoder_geo.get_feature_names_out(['Geography']))
geo_encode_df
```

```python
input_df = pd.DataFrame([input_data])
input_df
```

```python
input_df['Gender'] = label_encoder_gender.transform(input_df['Gender'])
input_df
```

```python
### Concatenation one hot encoded 

input_df= pd.concat([input_df.drop("Geography",axis=1),geo_encode_df],axis=1)
input_df
```

```python
input_scaled = scaler.transform(input_df)
input_scaled
```

```python
### Predict Churn 

prediction = model.predict(input_scaled)
prediction
```

```python
prediction_probability = prediction[0][0]
prediction_probability
```

```python
if prediction_probability > 0.5:
    print(f"Customer WILL churn  (Probability: {prediction_probability:.2%})")
else:
    print(f"Customer will NOT churn  (Probability: {prediction_probability:.2%})")
```
